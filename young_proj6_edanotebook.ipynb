# Step 7 Health Expectancy Dataset EDA  Objectives and Analysis Findings  

# Goal 1
Goal description: Determine average cost of each country to see who leds?

Hypothesis: USA will led since it one of the top 3 unhealthies countries and contributed data all 50 years  

Chart type: Bar chart

```python
# Calculate the average health expenditure for each country
average_costs = df.groupby('Country')['Spending_USD'].mean().sort_values(ascending=False)

# Plotting
plt.figure(figsize=(10, 6))
sns.barplot(x=average_costs.values, y=average_costs.index, palette='viridis')
plt.xlabel('Average Health Expenditure per Capita')
plt.ylabel('Country')
plt.title('Average Health Expenditure per Capita by Country')
plt.show()
```

# Conclusion

I was correct that USA would Led. We see the countries that more active in phyiscal terms spend less. 

# Goal 2
Goal description: Life exapantancy average vs highest/lowest country life expectancy

Hypothesis: You will see a correlation between more up to date year we get the higher our life expectancy becomes. Germany will have the highest and USA will have the lowest.

Chart type: Scarter plot 

```python
# Calculate the average life expectancy for each country for each year
average_life_expectancy =df.groupby(['Country', 'Year'])['Life_Expectancy'].mean().reset_index()

# Plotting
plt.figure(figsize=(10, 6))
sns.scatterplot(data=average_life_expectancy, x='Year', y='Life_Expectancy', hue='Country', palette='viridis')
plt.xlabel('Year')
plt.ylabel('Average Life Expectancy')
plt.title('Average Life Expectancy by Country and Year')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

# Conclusion

Hypothesis was wrong as Japan is the leader, but we do see that with all countries that over time thir is a steady increase in age. All countries saw a in life expectancy  drops in 2020 except for the leader China. 

# Goal 3
Goal description: Does increase speading on healthcare lead to higher life expectancy

Hypothesis: Yes I think it does

Chart type: Scatter

```python
# Calculate the average life expectancy and average health expenditure for each country
average_data = df.groupby('Country').agg({'Life_Expectancy': 'mean', 'Spending_USD': 'mean'}).reset_index()

# Calculate the average life expectancy and average health expenditure for each country for each year
average_data = df.groupby(['Year', 'Country']).agg({'Life_Expectancy': 'mean', 'Spending_USD': 'mean'}).reset_index()

# Plotting
plt.figure(figsize=(12, 8))
sns.scatterplot(data=average_data, x='Spending_USD', y='Life_Expectancy', hue='Country', style='Year', palette='viridis')
plt.xlabel('Average Health Expenditure per Capita USD')
plt.ylabel('Average Life Expectancy')
plt.title('Average Life Expectancy vs. Average Health Expenditure (by Year and Country)')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', title='Country')
plt.show()
```

# Conclusion

There is a direct link to increase life expectancy the more a person spends on it. There are some instances within countries where we see depending on year and the affects diease effections can skew the life expectancy numbers. Look at 2018 per covid vs 2020 when covid hit.
